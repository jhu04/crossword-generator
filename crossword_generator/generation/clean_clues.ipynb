```python
import wordninja

def split_words(s):
    contractions = ['d', 'm', 's', 't', 'll', 're', 've']
    split = wordninja.split(s)
    for i in range(len(split)-1, 0, -1):
        if split[i] in contractions:
            split[i-1] += split.pop(i)
    return ' '.join(split)

split_words('zees')
```

```python
import numpy as np
from tqdm import tqdm

def get_popularity(df, chunk_length=10, verbose=False, file_name='clean_clues.csv'):
    def popularity(terms: set[str], timeframe='today 5-y'):
        pytrend.build_payload(kw_list=terms)
        try:
            return int(pytrend.interest_over_time()[list(terms)].sum().sum())
        except KeyError:
            return 0
    
    path = os.path.join(const.DATA_ROOT, file_name)
    if not os.path.exists(path):
        with open(path, 'w') as create:
            pass
    
    try:
        start_i = len(pd.read_csv(path)) // chunk_length
    except pd.errors.EmptyDataError:
        start_i = 0
    chunks = np.array_split(df, len(df) // chunk_length)
    
    for i in tqdm(range(start_i, len(chunks))):
        chunks[i]['popularity'] = chunks[i][['answer', 'answer_sep']].values.tolist()
        chunks[i]['popularity'] = chunks[i]['popularity'].apply(lambda terms: popularity(set(terms)))
        if verbose:
            print(chunks[i])
        if i == 0:
            chunks[i].to_csv(path, mode='w', index=False)
        else:
            chunks[i].to_csv(path, mode='a', index=False, header=False)
```
